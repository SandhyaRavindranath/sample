```python
def data_extract(url):
    df = read_pdf(url,pages="all")
    for i in range(1, len(df)):
        df[0] = pd.concat([df[0],df[i]], axis=0)
    return df[0]
```

```python
pdf_files = [os.path.join(root, name)
             for root, dirs, files in os.walk(os.getcwd())
             for name in files
             if name.endswith((".pdf"))]
```

```python
def data_pdf(file):
    df = read_pdf(file, pages='all')
    list1 = [] 
    for item in df:
        for info in item.values:
            list1.append(info)
    df = pd.DataFrame(list1)
    return df
```
